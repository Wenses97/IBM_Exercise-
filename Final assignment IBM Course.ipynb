### The following code simulates a random match of rock, paper and scissors between two players.

```python
import random
options = ["rock", "paper", "scissors"]
p1 = options[random.randint(0,2)]
p2 = options[random.randint(0,2)]
if p1==p2:
    print("It's a tie!")
elif p1=="rock" and p2=="paper":
    print("Player 2 wins!")
elif p1=="rock" and p2=="scissors":
    print("Player 1 wins!")
elif p1=="paper" and p2=="rock":
    print("Player 1 wins!")
elif p1=="paper" and p2=="scissors":
    print("Player 2 wins!")
elif p1=="scissors" and p2=="paper":
    print("Player 1 wins!")
elif p1=="scissors" and p2=="rock":
    print("Player 2 wins!")
print("Player 1 chose: "+p1+". Player 2 chose: "+p2)
```
